# snake_ladder_monte_carlo_sim

```python
import random

def simulate_game():
    position = 1
    moves = 0
    snakes = {16: 6, 47: 26, 49: 11, 56: 53, 62: 19, 64: 60, 87: 24, 93: 73, 95: 75, 98: 78}
    ladders = {1: 38, 4: 14, 9: 31, 21: 42, 28: 84, 36: 44, 51: 67, 71: 91, 80: 100}

    while position < 100:
        roll = random.randint(1, 6)
        position += roll
        if position in snakes:
            position = snakes[position]
        elif position in ladders:
            position = ladders[position]
        moves += 1

    return moves

def monte_carlo_simulation(num_simulations):
    results = [simulate_game() for _ in range(num_simulations)]
    average_moves = sum(results) / num_simulations
    return average_moves

# Run the simulation 10,000 times
average = monte_carlo_simulation(10000)
print(f"Average number of moves to finish the game: {average_moves}")
```
